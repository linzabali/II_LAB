Построить распределение целевой переменной в виде гистограммы, сделать промежуточные выводы (обратите внимание на однородоность распределения и возможный дисбаланс). Посчитайте количество уникальных значений целевой переменной.

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'Downloads/data.csv' 
data = pd.read_csv(file_path)
target_variable = 'y' 

# Подсчет уникальных значений целевой переменной
unique_values = data[target_variable].nunique()
print(f"Количество уникальных значений целевой переменной: {unique_values}")

plt.figure(figsize=(10, 6))
data[target_variable].hist(bins=20) 
plt.title('Распределение целевой переменной')
plt.xlabel('Значения целевой переменной')
plt.ylabel('Частота')
plt.grid(False)
plt.show()

value_counts = data[target_variable].value_counts()
print("\nЧастота значений целевой переменной:")
print(value_counts)

# Проверка на однородность распределения и дисбаланс
if value_counts.values.max() > value_counts.values.mean() * 2:
    print("\nОбнаружен дисбаланс в распределении целевой переменной.")
else:
    print("\nДисбаланса в распределении не обнаружено.")
 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Downloads/data.csv')

var1 = 'Age'  
var2 = 'ReasonsForLack' 
target_var = 'y'   

if data[var1].dtype in ['int64', 'float64'] and data[var2].dtype in ['str']:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=data[var1], y=data[var2], hue=data[target_var], palette='viridis', alpha=0.7)
    plt.title(f'Scatter plot of {var1} vs {var2} colored by {target_var}')
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.legend(title=target_var)
    plt.show()
else:
    # Если переменные категориальные, используем heatmap
    contingency_table = pd.crosstab(data[var1], data[var2])
    plt.figure(figsize=(10, 6))
    sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlGnBu')
    plt.title(f'Heatmap of {var1} vs {var2}')
    plt.xlabel(var2)
    plt.ylabel(var1)
    plt.show()
```
